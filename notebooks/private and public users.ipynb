Load everything; libraries and data

```python
import pandas as pdusers_no_votes = pd.read_csv("../data/test/yelp_test_set_user.csv").drop_duplicates()users_with_votes = pd.read_csv("../data/train/yelp_academic_dataset_user.csv").drop_duplicates()users_all = users_with_votes.append(users_no_votes)reviews_train = pd.read_csv("../data/train/yelp_academic_dataset_review.csv").drop_duplicates()reviews_test = pd.read_csv("../data/test/yelp_test_set_review.csv").drop_duplicates()
```
